# Codenet

```python
from IPython.display import display
import matplotlib.pyplot as plt
import pdfkit as pk
import pandas as pd
import numpy as np
```

### Utils

```python
my_style = """
table {
  border-collapse: collapse;
}

th, td {
  text-align: left;
  padding: 8px;
}

tr:nth-child(even) {background-color: #f2f2f2;}
"""

def HTML_with_style(df):
    style = '<style>{}</style>'.format(my_style)
    df_html = df.to_html(index=False)

    return style + df_html
```

### Loading Datasets

```python
df = pd.read_csv("./metrics/codenet.csv", skipinitialspace=True)
cols = df.columns
```

### Obfuscation Techniques

#### ~> Table with the data

```python
# Obfuscation Techniques
selectedCols = ["te ollvm","tr+te ollvm","te ollvm_fla","tr+te ollvm_fla","te ollvm_sub","tr+te ollvm_sub","te ollvm_bcf","tr+te ollvm_bcf", "NO", "type"]
rf = df[df["model"] == "random_forest"][selectedCols]
mlp = df[df["model"] == "mlp"][selectedCols]
knn = df[df["model"] == "knn"][selectedCols]
reg_logistica = df[df["model"] == "reg_logistica"][selectedCols]
svm = df[df["model"] == "svm"][selectedCols]
sgd = df[df["model"] == "sgd"][selectedCols]

print("Random Forest")
display(rf)
print("MLP")
display(mlp)
print("KNN")
display(knn)
print("Reg Logística")
display(reg_logistica)
print("SVM")
display(svm)
print("SGD")
display(sgd)
```

#### ~> Bar Graph for each kind of combination of dataset

```python
fig, axs = plt.subplots(3,3,figsize=(12,10))

num=0
metric="acc"  
sample = df[df["type"] == metric][np.concatenate([selectedCols, ["model"]])]
for i in range(3):
    for k in range(3):
        sel = selectedCols[num]
        axs[i][k].set_title(sel)
        color = "b" if sel.startswith("te ") else "g"
        label = "Test Only" if sel.startswith("te ") else "Train+Test"
        axs[i][k].bar(range(sample["model"].size), sample[sel], color=color, label=label if i == 0 and k < 2 else "")
        axs[i][k].set_xticks(range(sample["model"].size), sample["model"], rotation=90)
        axs[i][k].set_ylim([0,1])
        
        num += 1

fig.suptitle(metric, fontsize=20)
fig.patch.set_facecolor('white')
fig.legend(loc='center left', bbox_to_anchor=(1, 0.5))
fig.tight_layout()
sample = df[df["type"] == metric][np.concatenate([selectedCols, ["model"]])]
sample = sample.loc[:, sample.columns != "type"].set_index("model").T
display(sample.mean())

# file_name="pdfs/teste.pdf"
# sample = sample.loc[:, (sample.columns != "type") & (sample.columns != "NO")].set_index("model").T
# sample = sample.mean()
# sample = pd.DataFrame({'Models':sample.index, 'Mean': sample.values})
# sample.columns = pd.MultiIndex.from_product([["Average accuracy with OLLVM in the POJ-104"], sample.columns])
# pk.from_string(HTML_with_style(sample), file_name)
```

```python
fig, ax = plt.subplots(1,1,figsize=(7,7))

metric="acc"  
sample = df[df["type"] == metric][np.concatenate([selectedCols, ["model"]])]


i, k = 0, 0
width = 0.4
sel = ["NO", "te ollvm", "tr+te ollvm", "te ollvm_sub", "tr+te ollvm_sub", "te ollvm_fla", "tr+te ollvm_fla"]
sel_labels = ["NOO\ntrain+test", "ollvm\ntest", "ollvm\ntrain+test", "ollvm_sub\ntest", "ollvm_sub\ntrain+test", "ollvm_fla\ntest", "ollvm_fla\ntrain+test"]
x = np.arange(len(sel))
alpha=0.6
plt.rcParams.update({'font.size': 14})

s = sample[sample["model"] == "random_forest"][np.concatenate([sel, ["model"]])]
ax.bar(x-width/2, s[sel].T[1], width=width, alpha=alpha, color="grey")

s = sample[sample["model"] == "mlp"][np.concatenate([sel, ["model"]])]
s = s[sel]
ax.bar(x+width/2, s[sel].T[5], width=width,alpha=alpha, color="green")

ax.set_title("TNP and TOP Models (Codenet)")
ax.set_xticks(x, sel_labels, rotation=60)
ax.set_ylim([0,1])
ax.set_ylabel("Accuracy")
ax.set_xlabel("Dataset and Phase")
ax.legend(["Random Forest", "MLP"])

# fig.patch.set_facecolor('white')
fig.tight_layout()
# plt.savefig("pdfs/accuracy.pdf")
# sample = df[df["type"] == "acc"][np.concatenate([selectedCols, ["model"]])]
# display(sample.mean())
```
