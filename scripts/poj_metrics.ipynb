```python
from IPython.display import display
import matplotlib.pyplot as plt
import pdfkit as pk
import pandas as pd
import numpy as np
```

### Utils

```python
my_style = """
table {
  border-collapse: collapse;
}

th, td {
  text-align: left;
  padding: 8px;
}

tr:nth-child(even) {background-color: #f2f2f2;}
"""

def HTML_with_style(df):
    style = '<style>{}</style>'.format(my_style)
    df_html = df.to_html(index=False)

    return style + df_html
```

### Loading Datasets

```python
df = pd.read_csv("./metrics/poj104.csv", skipinitialspace=True)
df2 = pd.read_csv("./metrics/data5.csv", skipinitialspace=True)
cols = df.columns
```

### Obfuscation Techniques

#### ~> Table with the data

```python
# Obfuscation Techniques
selectedCols = ["te ollvm","tr+te ollvm","te ollvm_fla","tr+te ollvm_fla","te ollvm_sub","tr+te ollvm_sub","te ollvm_bcf","tr+te ollvm_bcf", "NO", "type"]
rf = df[df["model"] == "random_forest"][selectedCols]
mlp = df[df["model"] == "mlp"][selectedCols]
knn = df[df["model"] == "knn"][selectedCols]
reg_logistica = df[df["model"] == "reg_logistica"][selectedCols]
svm = df[df["model"] == "svm"][selectedCols]
sgd = df[df["model"] == "sgd"][selectedCols]


rf2 = df2[df2["model"] == "random_forest"][selectedCols]
mlp2 = df2[df2["model"] == "mlp"][selectedCols]
knn2 = df2[df2["model"] == "knn"][selectedCols]
reg_logistica2 = df2[df2["model"] == "reg_logistica"][selectedCols]
svm2 = df2[df2["model"] == "svm"][selectedCols]
sgd2 = df2[df2["model"] == "sgd"][selectedCols]

print("Random Forest")
display(rf)
display(rf2)
print("MLP")
display(mlp)
display(mlp2)
print("KNN")
display(knn)
display(knn2)
print("Reg Logística")
display(reg_logistica)
display(reg_logistica2)
print("SVM")
display(svm)
display(svm2)
print("SGD")
display(sgd)
display(sgd2)
```

#### ~> Bar Graph for each kind of combination of dataset

```python
plot_five=True                           
fig, axs = plt.subplots(3,6 if plot_five else 3,figsize=(12,10))

num=0
metric="acc"  
sample = df[df["type"] == metric][np.concatenate([selectedCols, ["model"]])]
sample2 = df2[df2["type"] == metric][np.concatenate([selectedCols, ["model"]])]
for i in range(3):
    for j in range(3):
        k = j*2 if plot_five else j
        sel = selectedCols[num]
        axs[i][k].set_title(sel)
        color = "b" if sel.startswith("te ") else "g"
        label = "Test Only" if sel.startswith("te ") else "Train+Test"
        axs[i][k].bar(range(sample["model"].size), sample[sel], color=color, label=label if i == 0 and j < 2 else "")
        axs[i][k].set_xticks(range(sample["model"].size), sample["model"], rotation=90)
        axs[i][k].set_ylim([0,1])

        if(plot_five):
            axs[i][k+1].set_title(sel)
            axs[i][k+1].bar(range(sample2["model"].size), sample2[sel], color="purple")
            axs[i][k+1].set_xticks(range(sample2["model"].size), sample2["model"], rotation=90)
            axs[i][k+1].set_ylim([0,1])
        num += 1

fig.suptitle(metric, fontsize=20)
fig.patch.set_facecolor('white')
fig.legend(loc='center left', bbox_to_anchor=(1, 0.5))
fig.tight_layout()

sample = df[df["type"] == metric][np.concatenate([selectedCols, ["model"]])]
# sample = sample.loc[:, sample.columns != "type"].set_index("model").T
# display(sample.mean())

file_name="pdfs/mean_acc.pdf"
sample = sample.loc[:, (sample.columns != "type") & (sample.columns != "NO")].set_index("model").T
sample = sample.mean()
display(sample.index)
sample = pd.DataFrame({'Models':["KNN", "Random Forest", "SVM", "Logistic Regression", "SGD", "MLP"], 'Mean': sample.values})
sample.columns = pd.MultiIndex.from_product([["Average accuracy with OLLVM (POJ-104)"], sample.columns])
display(sample)
# pk.from_string(HTML_with_style(sample), file_name)
```
